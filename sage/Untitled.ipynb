```python
var('x y h E ν k n1 n2')
```

```python
G = E/2/(1+ν)
```

```python
w = 1/3*x^3*(x-1)^3*y^3*(y-1)^3-2*h^2/5/(1-ν)*(y^3*(y-1)^3*x*(x-1)*(5*x^2-5*x+1)+x^3*(x-1)^3*y*(y-1)*(5*y^2-5*y+1))
# w = 1/3*x^3*(x-1)^3*y^3*(y-1)^3
φ1 = y^3*(y-1)^3*x^2*(x-1)^2*(2*x-1)
φ2 = x^3*(x-1)^3*y^2*(y-1)^2*(2*y-1)
```

```python
κ11 = -derivative(φ1, x)
κ22 = -derivative(φ2, y)
κ12 = -0.5*(derivative(φ1, y) + derivative(φ2, x))
(κ11 + 2*x*(x-1)*(5*x^2-5*x+1)*y^3*(y-1)^3).simplify_full()
```

```python
γ1 = derivative(w, x) - φ1
γ2 = derivative(w, y) - φ2
(γ1+2*h^2/5/(1-ν)*(y^3*(y-1)^3*(20*x^3-30*x^2+12*x-1)+3*x^2*(x-1)^2*(2*x-1)*y*(y-1)*(5*y^2-5*y+1))).simplify_full()
```

```python
m11 = E*h^3/12/(1-ν^2)*(κ11 + ν*κ22)
m22 = E*h^3/12/(1-ν^2)*(ν*κ11 + κ22)
m12 = E*h^3/12/(1-ν^2)*(1-ν)*κ12
m11.factor())
```

```python
m1 = m11*n1 + m12*n2
m2 = m12*n1 + m22*n2
```

```python
q1 = k*G*h*γ1
q2 = k*G*h*γ2
```

```python
qn = q1*n1+q2*n2
```

```python
check_1 = derivative(m11, x) + derivative(m12, y) - q1
check_1.simplify_full()
```
